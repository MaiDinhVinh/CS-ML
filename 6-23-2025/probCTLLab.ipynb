```python
from matplotlib import pyplot as plt
from typing import List
import random
from collections import Counter
sample_mean_1 = []
for _ in range(100000):
    d = random.randint(1,6)
    sample_mean_1.append(d)
counter1 = Counter(sample_mean_1)

sample_mean_2 = []
for _ in range(100000):
    d = 0
    for _ in range(5):
        d += random.randint(1,6)
    dice = d/5
    sample_mean_2.append(dice)
counter2 = Counter(sample_mean_2)

sample_mean_3 = []
for _ in range(100000):
    d = 0
    for _ in range(30):
        d += random.randint(1,6)
    dice = d/30
    sample_mean_3.append(d)
counter3 = Counter(sample_mean_3)

fig, _ = plt.subplots(nrows = 1, ncols = 3, figsize=(30, 10))
fig.tight_layout(pad=10.0)

plt.subplot(1,3,1)
plt.bar(counter1.keys(), counter1.values())
plt.xticks(size=20)
plt.yticks(size=20)
plt.xlabel("Score",size=20)
plt.ylabel("Count",size=20)
plt.title("Histogram - Throw 1 dice 100k times",size=20)

plt.subplot(1,3,2)
plt.bar(counter2.keys(), [count/100000 for count in counter2.values()])
plt.xlabel("Mean", size=20)
plt.ylabel("Prob", size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.title("PMF - Throw 5 dices 100k times", size=20)

plt.subplot(1,3,3)
plt.bar(counter3.keys(), [count/100000 for count in counter3.values()])
plt.xlabel("Mean", size=20)
plt.ylabel("Prob", size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.title("PMF - Throw 30 dices 100k times", size=20)

plt.show()
```
